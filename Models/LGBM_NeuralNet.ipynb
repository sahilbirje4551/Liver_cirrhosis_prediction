```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
```

```python
train = pd.read_csv(r"C:\Users\sahil\Downloads\GDSC\train.csv")
```

```python
train = train.drop(['id'], axis = 1)
```

```python
categorical_cols = list(train.select_dtypes(include=['object']).columns)
print(categorical_cols)
```

```python
numerical_cols = list(train.select_dtypes(include=['int64','float64']).columns)
print(numerical_cols)
```

```python
#excluding Edema we can label encode these categorical cols
categorical_cols.remove('Edema')
print(categorical_cols)
```

```python
label_encoder = LabelEncoder()
for i in categorical_cols:
    train[i] = label_encoder.fit_transform(train[i])

display(train)
```

```python
X = train.drop(['Status'], axis = 1)
y = train['Status']
```

```python
X = pd.get_dummies(X, columns = ['Edema'])
```

```python
X
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

```python
# Normalization
scaler = MinMaxScaler()
X = scaler.fit_transform(X)
print("\nNormalized data:")
print(X)
```

```python
import lightgbm as lgb
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=584)

# Balancing the training examples

from imblearn.over_sampling import RandomOverSampler

sampler = RandomOverSampler(random_state = 767)
X_train_resampled, y_train_resampled = sampler.fit_resample(X_train, y_train)

train_data = lgb.Dataset(X_train_resampled, label=y_train_resampled)
test_data = lgb.Dataset(X_test, label=y_test)

# Set parameters for LightGBM for multiclass classification
params = {
    'objective': 'multiclass',
    'num_class': 3,
    'metric': 'multi_error',
    'boosting_type': 'gbdt',
    'num_leaves': 31,
    'learning_rate': 0.05,
    'feature_fraction': 0.9
}

# Train the model
num_round = 500
bst = lgb.train(params, train_data, num_round, valid_sets=[test_data], early_stopping_rounds=10)

# class_names = [0, 1, 2]
class_names = ['C', 'CL', 'D']
```

```python
y_train_proba = bst.predict(X_train, num_iteration=bst.best_iteration)
y_pred_proba = y_train_proba
```

```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import ModelCheckpoint
from tensorflow import keras

# Define the neural network architecture
model = Sequential()
model.add(Dense(64, activation='relu', input_dim=y_pred_proba.shape[1]))
model.add(Dense(128, activation='relu'))
model.add(Dense(64, activation='relu'))
model.add(Dense(64, activation='relu'))
model.add(Dense(128, activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(64, activation='relu'))
model.add(Dense(128, activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(3, activation='softmax'))  

model.summary()

# Compile the model
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Define a ModelCheckpoint callback to save the best model
checkpoint = ModelCheckpoint('best_model_1.h5', monitor='val_accuracy', verbose=1, save_best_only=True, mode='max')

# Train the model with ModelCheckpoint callback
history = model.fit(y_pred_proba, y_train, epochs=75, batch_size=4, validation_split=0.5, callbacks=[checkpoint])

# Load the best model for prediction
best_model = keras.models.load_model('best_model_1.h5')
```

```python
y_test_proba = bst.predict(X_test, num_iteration=bst.best_iteration)
y_test_pred_proba = y_test_proba
```

```python
# Evaluate the best model on the test set
loss, accuracy = best_model.evaluate(y_test_pred_proba, y_test)
print("Test Accuracy:", accuracy)

# Predict using the best model
y_pred = best_model.predict(y_test_pred_proba)
```

```python
from sklearn.metrics import classification_report

# y_pred = model.predict(y_test_pred_proba)
report = classification_report(y_test, np.argmax(y_pred, axis=1))

print("Classification Report:")
print(report)
```
